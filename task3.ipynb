# 기상 데이터 /특보데이터 EDA

```python
fire2
```

```python
fire2['fire_weather.district_1'].unique()
# select only rows where district_1 is '서울'
fire2_seoul=fire2[fire2['fire_weather.district_1']=='서울특별시']
```

```python
fire2_seoul
# for every day, plot temperature
#fire2_seoul['fire_weather.tm'] = pd.to_datetime(fire2_seoul['fire_weather.tm'])

# for every day, get the average temperature
fire_day_avg_temp=fire2_seoul.groupby('fire_weather.tm')['fire_weather.ta_max'].mean()
fire_day_var_temp=fire2_seoul.groupby('fire_weather.tm')['fire_weather.ta_max'].var()
```

```python
fire1_seoul=fire1[fire1['fire_firefighting.district_1']=='서울특별시']
```

```python
fire1_seoul['date'] = pd.to_datetime(fire1_seoul['fire_firefighting.tm']).dt.date
```

```python
# for every day, get the count of fires
fire_day_count=fire1_seoul.groupby('date').size()
```

```python
fire_day_count
# change dtype to float
fire_day_count=fire_day_count.astype(float)
```

```python
fire_day_count
```

```python
fire_day_avg_temp
#change columnname to 'date'
fire_day_avg_temp=fire_day_avg_temp.reset_index()
```

```python
fire_day_count=fire_day_count.reset_index()
```

```python
fire_day_avg_temp.rename(columns={'fire_weather.tm':'date', 'fire_weather.ta_max':'avg_temp'}, inplace=True)
```

```python
fire_day_counts=np.array(fire_day_count[0])
fire_day_avg_temps=np.array(fire_day_avg_temp['avg_temp'])
#plt.plot(fire_day_count['date'], fire_day_count[0], label='fire_count',color='red') 
```

```python
# minmax scaling for fire count
fire_day_counts=(fire_day_counts-np.min(fire_day_counts))/(np.max(fire_day_counts)-np.min(fire_day_counts))
fire_day_avg_temps=(fire_day_avg_temps-np.min(fire_day_avg_temps))/(np.max(fire_day_avg_temps)-np.min(fire_day_avg_temps))
```

```python
plt.plot(fire_day_counts, label='fire_count',color='red')
plt.plot(fire_day_avg_temps, label='avg_temp',color='blue')

plt.title('서울특별시 일별 화재발생건수와 평균기온')
plt.xlabel('날짜')
plt.ylabel('건수/기온')
plt.legend()
```

# correlation between temperature, fire count in seoul

```python
plt.scatter(fire_day_counts,fire_day_avg_temps, label='fire_count',color='red')
```

```python
# for every day, get the average humidity
fire_day_avg_humid=fire2_seoul.groupby('fire_weather.tm')['fire_weather.hm_mean'].mean()
fire_day_var_humid=fire2_seoul.groupby('fire_weather.tm')['fire_weather.hm_mean'].var()
```

```python
fire_day_avg_humid
```

```python
# do the same thing , compare fire count with humidity
fire_day_avg_humid=fire_day_avg_humid.reset_index()


fire_day_avg_humids=np.array(fire_day_avg_humid['fire_weather.hm_mean'])

# minmax scaling for humid
fire_day_avg_humids=(fire_day_avg_humids-np.min(fire_day_avg_humids))/(np.max(fire_day_avg_humids)-np.min(fire_day_avg_humids))

```

```python
plt.plot(fire_day_avg_humids, label='avg_humid',color='green')
plt.plot(fire_day_counts, label='fire_count',color='red')

plt.title('서울특별시 일별 화재발생건수와 평균습도')
plt.xlabel('날짜')
plt.ylabel('건수/습도')
plt.legend()
```

```python
plt.scatter(fire_day_counts,fire_day_avg_humids, label='fire_count',color='red')
# GET CORR
print("화제 발생건수와 평균기온의 상관계수")
print(np.corrcoef(fire_day_counts,fire_day_avg_temps))
print("화제 발생건수와 평균습도의 상관계수")
print(np.corrcoef(fire_day_counts,fire_day_avg_humids))
```
